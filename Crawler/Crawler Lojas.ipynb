```python
domain = ["https://www.boutiquedassi.com.br","https://www.lojasexclusiva.com.br",
         "https://www.bessaloja.com.br", "https://www.bluk.com.br",
         "https://www.achados96.com.br", "https://www.babyminidiva.com.br",
         "https://www.bagy.com.br/maroquinhasbrecho","https://www.lojavillevie.com.br",
         "https://www.latiendaa.com.br/", "https://www.bellaseda.online"]
```

```python
def bsfCrawl(links,maxPages):
    count= 0
    linkArray = [links] #array que ter√° todos links de uma loja
#     seen = False
    while( count < maxPages):
        print("entrei no loop")
        codigo_fonte = rq.get(links)
        soup = BeautifulSoup(codigo_fonte.text,'lxml')
        for link in soup.find_all('a', href=True):
            href = link["href"]
            linkArray.append(href)
            print (href)
        count=count+1
    return linkArray
```
